# Obteniendo los datos

```python
def obtener_data(ruta):
    df = pd.read_json(ruta, lines=True)
    return df
```

```python
ruta = 'tweets/farmers-protest-tweets-2021-03-5.json'
df = obtener_data(ruta)
print(df)
```

# Obteniendo los 10 mas retweeteados

```python
def mas_retweeted(data):
    retweet = data.sort_values(by ='retweetCount', axis = 0, ascending = False)
    return retweet.head(n=10)
```

# Obteniendo a los usuarios con m√°s tweets

```python
def usuarios_tweeteros(data):
    column_2 = data['user']
    columna_3 = column_2.apply(lambda row: row['id'])
    columna_4 = columna_3.value_counts()
    column_6 = columna_4.head(n=10).index
    lista = []
    for i in range(len(column_6)):
        for j in range(len(column_2)):
            if int(column_2[j]['id']) == int(column_6[i]):
                lista.append(column_2[j])
                break
    dataframe = pd.DataFrame(lista)
    return dataframe
```

# Obteniendo los 10 d√≠as donde hay m√°s tweets

```python
def dias_mas_tweeteados(data):
    day = pd.to_datetime(data['date']).dt.date
    day_2 = day.value_counts()
    return day_2.head(n=10)
```
